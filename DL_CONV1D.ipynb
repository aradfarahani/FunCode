### Pre Processing

```python
#normalizing
Data_file = (my_file - my_file.min() + 1)/(my_file.max() - my_file.min() + 1)
```

```python
print(Data_file.head())
print(len(Data_file))
```

### Deviding Dataset

```python
from numpy import random
np.random.seed(42)
#Data_file_len = len(Data_file)
Data_file_len = 896
t = list((range(Data_file_len)))
split = int(np.floor(0.1*Data_file_len))
random.shuffle(t)
train_data_indices, test_data_indices = t[split:], t[:split]
```

```python
Data_file = Data_file.to_numpy()
```

```python
train_data = Data_file[train_data_indices]
test_data = Data_file[test_data_indices]
```

```python
print(((train_data)))
```

```python
input_train = train_data[0:, 0:4]
target_train = train_data[0:, 4:5]
input_test = test_data[0:, 0:4]
target_test = test_data[0:, 4:5]
```

```python
x_training = input_train.reshape(input_train.shape[0], 4, 1)
y_training = target_train.reshape(target_train.shape[0], 1, 1)
x_test = input_test.reshape(input_test.shape[0], 4, 1)
y_test = target_test.reshape(input_test.shape[0], 1, 1)
```

```python
print(x_training.shape)
print(y_test.shape)
print(y_test[10, 0])
```

### Creating Model

```python
from keras import layers
model = tf.keras.Sequential()

activation_func = 'relu'

model.add(layers.Conv1D(filters = 64, kernel_size = 5, strides = 1, padding = 'same', activation = activation_func, input_shape = (4, 1), name="Conv1D_1"))
model.add(Dropout(0.50))
model.add(layers.Conv1D(filters = 32, kernel_size = 3, strides = 1, padding = 'same', activation = activation_func, name="Conv1D_2"))
#model.add(layers.MaxPooling1D(pool_size = 2))

model.add(layers.Conv1D(filters = 16, kernel_size = 2, strides = 1, padding = 'same', activation = activation_func, name="Conv1D_3"))
model.add(layers.Conv1D(filters = 16, kernel_size = 2, strides = 1, padding = 'same', activation = activation_func, name="Conv1D_4"))
model.add(layers.Conv1D(filters = 16, kernel_size = 2, strides = 1, padding = 'same', activation = activation_func, name="Conv1D_5"))
model.add(layers.MaxPooling1D(pool_size=2 , name="MaxPooling1D_1"))

# model.add(layers.Conv1D(filters = 32, kernel_size = 7, strides = 1, padding = 'same', activation = activation_func, name="Conv1D_4"))
# model.add(layers.Conv1D(filters = 64, kernel_size = 7, strides = 1, padding = 'same', activation = activation_func, name="Conv1D_5"))
# model.add(layers.Conv1D(filters = 128, kernel_size = 7, strides = 1, padding = 'same', activation = activation_func, name="Conv1D_6"))
# model.add(layers.MaxPooling1D(pool_size=5 , name="MaxPooling1D_2"))
# conv 2
# model.add(layers.Conv1D(filters = 128, kernel_size = 2, strides = 1, padding = 'same', activation = activation_func))
# model.add(layers.Conv1D(filters = 64, kernel_size = 2, strides = 1, padding = 'same', activation = activation_func))
#model.add(layers.MaxPooling1D(pool_size = 3))

#model.add(GlobalAveragePooling1D())

model.add(Flatten())
model.add(layers.Dense(32, activation = 'relu'))
#model.add(layers.Dense(300, activation = 'relu'))
#last layer
model.add(layers.Dense(1,activation = 'linear'))

model.summary()
```

#### Compiling the Model

```python
loss_func = 'mse'
optimizer_func = tf.keras.optimizers.RMSprop(0.001)
#optimizer_func = tf.keras.optimizers.SGD(learning_rate = 0.0001)
# optimizer_func = tf.keras.optimizers.Adagrad(learning_rate = 0.1)

model.compile(loss=loss_func,
              optimizer=optimizer_func,
              metrics=['RootMeanSquaredError', 'mse','mae'])
```

#### Training the Model

```python
import datetime
start = datetime.datetime.now()
trainedModel = model.fit(x_training, y_training, batch_size = 256, epochs= 510, verbose = 1, validation_split = 0.15, shuffle = True)
end = datetime.datetime.now()
```

```python
print('Training Duration: ', end-start)
```

#### Evaluating the Model

```python
history = trainedModel.history
rmse = history['RootMeanSquaredError']
validation_rmse = history['val_RootMeanSquaredError']
mae = history['mae']
validation_mae = history['val_mae']
mse = history['mse']
validation_mse = history['val_mse']


plt.plot(mae)
plt.plot(validation_mae)
plt.legend(['Mean Absolute Error','validation_Mean Absolute Error'])
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.figure()
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.plot(mse)
plt.plot(validation_mse)
plt.legend(['Mean Squared Error','validation_Mean Squared Error'])
plt.figure()
plt.xlabel('Epochs')
plt.ylabel("Error")
plt.plot(rmse)
plt.plot(validation_rmse)
plt.legend(['Root Mean Squared Error', 'validation_Root Mean Squared Error'])
```
