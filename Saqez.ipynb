```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import pandas as pd

df = pd.read_csv("Saqez.csv")

center_x = df['x'].mean()
center_y = df['y'].mean()
radius = 6000 

mask = (df['x'] - center_x)**2 + (df['y'] - center_y)**2 < radius**2

selected_points = df[mask]
selected_x = selected_points['x']
selected_y = selected_points['y']
selected_fac1 = selected_points['fac1']

fig, ax = plt.subplots(figsize=(8, 8)) 

ax.scatter(df['x'], df['y'], c='lightgray')  # Plot all points in light gray
circle = Circle((center_x, center_y), radius, color='red', fill=False)
ax.add_patch(circle)

scatter = ax.scatter(selected_x, selected_y, c=selected_fac1, cmap='jet')

plt.colorbar(scatter, label='Factor 1') 
plt.title('Scatter Plot of x and y Colored by Factor 1')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(rotation=45, fontsize='small')
plt.yticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import pandas as pd

df = pd.read_csv("Saqez.csv")

center_x = df['x'].mean()
center_y = df['y'].mean()
radius = 6000  

mask = (df['x'] - center_x)**2 + (df['y'] - center_y)**2 < radius**2

selected_points = df[mask]
selected_x = selected_points['x']
selected_y = selected_points['y']
selected_fac1 = selected_points['fac1']

fig, ax = plt.subplots(figsize=(10, 10)) 

ax.scatter(df['x'], df['y'], c='lightgray') 
circle = Circle((center_x, center_y), radius, color='red', fill=False)
ax.add_patch(circle)

scatter = ax.scatter(selected_x, selected_y, c=selected_fac1, cmap='jet')

ax.set_aspect('equal', 'box')

plt.colorbar(scatter, label='Factor 1') 
plt.title('Scatter Plot of x and y Colored by Factor 1')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(rotation=45, fontsize='small')
plt.yticks(rotation=45)

plt.show()
```
