```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d


path = "Book11.xlsx"
df = pd.read_excel(path)

d = df.iloc[:,1].values
passing = df.iloc[:,2].values  

sort_idx = np.argsort(d)
d = d[sort_idx]
passing = passing[sort_idx]


f = interp1d(passing, d, fill_value="extrapolate")

D10 = float(f(10))
D30 = float(f(30))
D60 = float(f(60))

Cu = D60 / D10
Cc = (D30**2) / (D60 * D10)

print("D10 =", D10)
print("D30 =", D30)
print("D60 =", D60)
print("Cu  =", Cu)
print("Cc  =", Cc)


plt.figure(figsize=(8,6))
plt.semilogx(d, passing, marker='o', linewidth=1.5)

plt.gca().invert_xaxis()

plt.semilogx([D10, D30, D60], [10, 30, 60], 'o', markersize=8)

plt.text(D10, 10, f" D10={D10:.3f} mm")
plt.text(D30, 30, f" D30={D30:.3f} mm")
plt.text(D60, 60, f" D60={D60:.3f} mm")

plt.xlabel("Particle size (mm)")
plt.ylabel("Percent Passing (%)")
plt.title("Sieve Curve")

plt.grid(True, which="both", linestyle="--", linewidth=0.5)

plt.ylim(0, 100)
plt.xlim(max(d)*1.1, min(d)/1.1)

plt.tight_layout()
plt.show()
```
