```python
import rasterio
import matplotlib.pyplot as plt

# Path to your DEM
dem_path = "DEM_ISFAHAN.dem"

# Open with rasterio
with rasterio.open(dem_path) as src:
    dem = src.read(1)             # read first (and only) band
    profile = src.profile
    bounds = src.bounds

print("CRS:", profile["crs"])
print("Shape:", dem.shape)
print("Bounds:", bounds)

# Mask NoData values
dem = dem.astype(float)
dem[dem == src.nodata] = float("nan")

# Plot DEM
plt.figure(figsize=(10,8))
plt.imshow(dem, cmap="terrain", extent=[bounds.left, bounds.right, bounds.bottom, bounds.top])
plt.colorbar(label="Elevation (m)")
plt.title("DEM - Isfahan")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
