```python
import numpy as np
import matplotlib.pyplot as plt

a, b, n = 0, np.pi / 2, 10

def f(x):
    return np.sin(x)

fig, axes = plt.subplots(2, 5, figsize=(18, 7))
axes = axes.ravel()

for idx, n in enumerate(range(1, 11)):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    I = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])

    x_smooth = np.linspace(a, b, 1000)
    y_smooth = f(x_smooth)
    axes[idx].plot(x_smooth, y_smooth, 'b', label='f(x)=sin(x)')
    axes[idx].fill_between(x_smooth, 0, y_smooth, color='skyblue', alpha=0.3)

    for i in range(n):
        xs = [x[i], x[i], x[i+1], x[i+1]]
        ys = [0, y[i], y[i+1], 0]
        axes[idx].fill(xs, ys, 'orange', edgecolor='black', alpha=0.4)

    axes[idx].set_title(f"n={n},  S ≈ {I:.7f}")
    axes[idx].set_xlim(a, b)
    axes[idx].set_ylim(0, 1.1)
    axes[idx].grid(True)

plt.show()
```
